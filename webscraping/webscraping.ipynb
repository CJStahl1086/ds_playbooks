```python
import pandas as pd
import numpy as np
```

```python
#Scraping Conference and Division information of NFL teams
all_tables = pd.read_html('https://en.wikipedia.org/wiki/National_Football_League')

conf = all_tables[3]
conf = conf[['Division[66]', 'Club[66]']]
afc = conf.iloc[0:16].copy()
nfc = conf.iloc[17:33].copy()

nfc.columns = nfc.columns.droplevel()
afc.columns = afc.columns.droplevel()

afc.columns.values[1] = 'Team'
afc.columns.values[0] = 'Division'
afc['Conference'] = 'AFC'

nfc.columns.values[1] = 'Team'
nfc.columns.values[0] = 'Division'
nfc['Conference'] = 'NFC'

nfl = pd.concat([afc, nfc], axis=0)
nfl['Team'] = nfl['Team'].str.replace('*','')
nfl['Team'] = nfl['Team'].str.replace('â€ ','')
nfl.head(5)
```

```python
#Scraping Playoff Years of Individual NFL Teams
playoffs = pd.read_html('https://en.wikipedia.org/wiki/List_of_NFL_playoff_games')
AZ = playoffs[0]
AZyears = np.array(AZ['Occurrences'])
AZyears = ','.join(str(x) for x in AZyears)
AZyears
```

```python
#Merging Playoff years with previous dataframe
nfl['Playoff_Years'] = ''
nfl['Playoff_Years'] = np.where(nfl['Team'] == 'Arizona Cardinals', AZyears, nfl['Playoff_Years'])
nfl
```
