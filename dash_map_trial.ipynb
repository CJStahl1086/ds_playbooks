```python
import plotly.graph_objs as go
import plotly.offline as pyo
from plotly.subplots import make_subplots
import pandas as pd

# Create data frame
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Sales': [100, 120, 150, 170, 200],
        'Expenses': [70, 90, 100, 120, 150],
        'Profit': [30, 30, 50, 50, 50]}
df = pd.DataFrame(data)

# Create line chart
line = go.Scatter(
    x=df['Year'],
    y=df['Sales'],
    mode='lines+markers',
    name='Sales',
    line=dict(color='white')
)

# Create bar chart
bar = go.Bar(
    x=df['Year'],
    y=df['Expenses'],
    name='Expenses',
    marker=dict(color='white')
)

# Create scatter chart
scatter = go.Scatter(
    x=df['Year'],
    y=df['Profit'],
    mode='markers',
    name='Profit',
    marker=dict(color='white', size=10)
)

# Create subplot grid
fig = make_subplots(rows=2, cols=2, 
                    subplot_titles=('Sales', 'Expenses', 'Profit'),
                    specs=[[{'rowspan': 2}, {}], [None, {}]])

# Add line chart to subplot 1
fig.add_trace(line, row=1, col=1)

# Add bar chart to subplot 2
fig.add_trace(bar, row=1, col=2)

# Add scatter chart to subplot 3
fig.add_trace(scatter, row=2, col=2)

# Define layout
fig.update_layout(
    title='Dashboard',
    xaxis=dict(title='Year', color='white'),
    yaxis=dict(title='Amount', color='white'),
    plot_bgcolor='black',
    paper_bgcolor='black'
)

# Save dashboard to HTML file
pyo.plot(fig, filename='dashboard.html')
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo
from plotly.subplots import make_subplots
import pandas as pd

# Create data frame
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Sales': [100, 120, 150, 170, 200],
        'Expenses': [70, 90, 100, 120, 150],
        'Profit': [30, 30, 50, 50, 50]}
df = pd.DataFrame(data)

# Create line chart
line = go.Scatter(
    x=df['Year'],
    y=df['Sales'],
    mode='lines+markers',
    name='Sales',
    line=dict(color='white')
)

# Create bar chart
bar = go.Bar(
    x=df['Year'],
    y=df['Expenses'],
    name='Expenses',
    marker=dict(color='white')
)

# Create scatter chart
scatter = go.Scatter(
    x=df['Year'],
    y=df['Profit'],
    mode='markers',
    name='Profit',
    marker=dict(color='white', size=10)
)

# Create subplot grid
fig = make_subplots(rows=2, cols=2, 
                    subplot_titles=('Sales', 'Expenses', 'Profit'),
                    specs=[[{'rowspan': 2}, {}], [None, {}]])

# Add line chart to subplot 1
fig.add_trace(line, row=1, col=1)

# Add bar chart to subplot 2
fig.add_trace(bar, row=1, col=2)

# Add scatter chart to subplot 3
fig.add_trace(scatter, row=2, col=2)

# Define layout
fig.update_layout(
    title='Dashboard',
    xaxis=dict(title='Year', color='blue'),
    yaxis=dict(title='Amount', color='blue', showgrid=True, gridcolor='white', zerolinecolor='blue'),
    plot_bgcolor='black',
    paper_bgcolor='black'
)

# Save dashboard to HTML file
pyo.plot(fig, filename='dashboard.html')
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo

# Define map data
map_data = [
    go.Scattermapbox(
        lat=['46.9'],
        lon=['-110.4'],
        mode='markers',
        marker=dict(
            size=14,
            color='white'
        ),
        text=['Montana'],
        hoverinfo='text'
    )
]

# Define layout
layout = go.Layout(
    title='Map Dashboard',
    mapbox=dict(
        style='stamen-terrain',
        center=dict(
            lat=46.9,
            lon=-110.4
        ),
        zoom=4
    ),
    plot_bgcolor='black',
    paper_bgcolor='black'
)

# Create figure
fig = go.Figure(data=map_data, layout=layout)

# Save dashboard to HTML file
pyo.plot(fig, filename='dashboard.html')
```
