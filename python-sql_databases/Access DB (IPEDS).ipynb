```python
import pandas as pd
import pyodbc
```

```python
# CONNECTION TO ACCESS DB VIA PYODBC
conn_str = (r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
            r'DBQ=C:\Users\chase\Desktop\IPEDS202021.accdb;')
conn = pyodbc.connect(conn_str)
```

```python
# RETRIEVING ENROLLMENT BY INSTITUTION
a = pd.read_sql(sql="SELECT * FROM Ef2020", con=conn)
a = a.groupby('UNITID')[['EFTOTAL']].sum().reset_index()

# RETRIEVING STATE OF EACH INSTITUTION
b = pd.read_sql(sql="SELECT * FROM HD2020", con=conn)
b = b[['UNITID','STABBR']]

# MERGING TO CREATE TOTAL ENROLLMENT BY STATE
final = pd.merge(a, b, on='UNITID')
final = pd.DataFrame(final.groupby('STABBR')['EFTOTAL'].sum()).reset_index()
final.to_csv('final.csv')
```
